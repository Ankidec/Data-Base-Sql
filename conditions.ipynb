```sql
SELECT
   title,
   length,
   CASE
      WHEN length<=60 THEN 'short movie'
      WHEN length<=90 THEN 'normal movie'
      WHEN length<=120 THEN 'long movie'
      ELSE 'extra long movie'
   END as movie_length
   FROM
     film;           
```

```sql
SELECT
   title,
   length,
   CASE
      WHEN length<=60 THEN 'short movie'
      WHEN length<=90 THEN 'normal movie'
      WHEN length<=120 THEN 'long movie'
      ELSE 'extra long movie'
   END as movie_length
   FROM
     film
  ORDER BY 
     length;   
```

```sql
SELECT
   CASE
      WHEN length<=60 THEN 'short movie'
      WHEN length<=90 THEN 'normal movie'
      WHEN length<=120 THEN 'long movie'
      ELSE 'extra long movie'
   END as movie_length,
   COUNT(1) as number_of_movies
   FROM
     film
  GROUP BY movie_length;
```
